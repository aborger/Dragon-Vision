```python
import pathlib
import shutil
```

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE
```

```python
# Make folder for pics
directory = "Dragon_Pics"
class_name = "People"
# Parent Directories 
parent_dir = "/Users/borge/.keras/datasets"
  
# Path 
PATH = os.path.join(parent_dir, directory) 
  
# Create the directory 
os.makedirs(PATH, exist_ok = True) 

# Create class in directory
class_dir = PATH
class_dir = os.path.join(class_dir, class_name)
os.makedirs(class_dir, exist_ok = True)
```

```python
# collect urls, will read from file
urls = ['https://c7.staticflickr.com/6/5499/10245691204_98dce75b5a_o.jpg',
       'https://farm1.staticflickr.com/5615/15335861457_ec2be7a54e_o.jpg',
       'https://c7.staticflickr.com/8/7590/17048042861_97168daff8_o.jpg',
       'https://farm5.staticflickr.com/5582/18233009494_029b52ca79_o.jpg',
       'https://farm6.staticflickr.com/4126/5145819744_b4a7871064_o.jpg']
```

```python
# download each url and place in folder
for u in range(0, len(urls)):
    data_dir = tf.keras.utils.get_file(origin=urls[u],
                                  fname='pic'+str(u)+'.jpg')
    data_dir = pathlib.Path(data_dir)
    shutil.move(str(data_dir), str(class_dir))
```

```python
num = len(os.listdir(PATH))
print(num)
```

```python
# Generator for training data
train_image_generator = ImageDataGenerator(rescale=1./255)
```

```python
# variables to use for preprocessing
batch_size = 128
epochs = 15
IMG_HEIGHT = 150
IMG_WIDTH = 150
```

```python
train_data_gen = train_image_generator.flow_from_directory(batch_size=batch_size,
                                                          directory = PATH,
                                                          shuffle = True,
                                                          target_size=(IMG_HEIGHT, IMG_WIDTH),
                                                           class_mode='binary')
```

```python
training_images, _ = next(train_data_gen)
```

```python
# This function will plot images in the form of a grid with 1 row and 5 columns where images are placed in each column.
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
plotImages(training_images[:5])
```
